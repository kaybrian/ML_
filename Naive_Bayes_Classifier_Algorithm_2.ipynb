```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/spam.csv')
```

```python
data.head()
```

```python
data.groupby('Category').describe()
```

```python
# convert the catergory into 1 and 0
data['spam'] = data['Category'].apply(lambda x: 1 if x=='spam' else 0)
```

```python
data.head()
```

```python
# get the train test splits
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data['Message'], data['spam'], test_size=0.25, random_state=101)
```

```python
# change the message into numbers
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X_train_count = cv.fit_transform(X_train.values)
X_train_count.toarray()[:3]
```

```python
#  using multinomialNB
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count, y_train)
```

```python
# get tow emails
email = [
    'Hey Brian, how are you doing today?, can we get some time to watch a match together tomorrow? ',
    'upto 20% discount on your first order, exclusive offers for you. Dont miss this reward!'
]

# make a prediction for the emails
email_count = cv.transform(email)
model.predict(email_count)
```

```python
#  make a prediction and score of the model
X_test_count = cv.transform(X_test)
model.score(X_test_count, y_test)
```

```python
# using the sklearn pipeline
from sklearn.pipeline import Pipeline
clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())
])

# fit the model
clf.fit(X_train, y_train)

# make the prediction
clf.predict(email)
```

```python
# get the perfromace of the classifier
from sklearn.metrics import classification_report
print(classification_report(y_test, clf.predict(X_test)))
```
