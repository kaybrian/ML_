```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
#
```

```python
dataset = pd.read_csv('/content/Melbourne_housing_FULL.csv')

dataset.head()
```

```python
dataset.shape
```

```python
dataset.info()
```

```python
cols_to_use = ['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount',
               'Distance', 'CouncilArea', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']
dataset = dataset[cols_to_use]
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
# check for the nan values
dataset.isna().sum()
```

```python
cols_to_fill_zero = ['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']
dataset[cols_to_fill_zero] = dataset[cols_to_fill_zero].fillna(0)
```

```python
# fill the Nan values with the mean
dataset['Landsize'] = dataset['Landsize'].fillna(dataset.Landsize.mean())
dataset['BuildingArea'] = dataset['BuildingArea'].fillna(dataset.BuildingArea.mean())
```

```python
# check for the nan values again
dataset.isna().sum()
```

```python
dataset.dropna(inplace=True)
```

```python
dataset.shape
```

```python
# missing values
34857 - 27244
```

```python
dataset = pd.get_dummies(dataset, drop_first=True)
```

```python
dataset
```

```python
#  split the data in to X and y for the dataset
X = dataset.drop('Price', axis=1)
y = dataset['Price']
```

```python
# make sure all the Trues and Falses in each row and column is either 1 or 0
```

```python
# replace any True value to 1 and false to 0
X.replace({True: 1, False: 0}, inplace=True)
```

```python
X.head()
```

```python
y.head()
```

```python
# do the train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2)
```

```python
# linear regression
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_train, y_train)
lr_clf.score(X_test, y_test)
```

```python
# score on training dataset
# score on testing dataset
lr_clf.score(X_train, y_train)
```

```python
#  prefect example of overfitting in the data here

# lets add the L1 or L2 regularization
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
```

```python
lesso_reg = Lasso(alpha=50, max_iter=100, tol=0.1)
lesso_reg.fit(X_train, y_train)
lesso_reg.score(X_test, y_test)
```

```python
# using the Ridge regularization
ridge_reg = Ridge(alpha=50)
ridge_reg.fit(X_train, y_train)
ridge_reg.score(X_test, y_test)
```

```python
# compare the test scores
lesso_reg.score(X_test, y_test)
```

```python
# compare with the tests of ridge
ridge_reg.score(X_test, y_test)
```
